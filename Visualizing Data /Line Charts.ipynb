```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook
```

```python
variance     = [1,2,4,8,16,32,65,128,256]
bias_squared = [256,128,64,32,16,8,4,2,1]

total_erro   = [x+y for x ,y in zip(variance,bias_squared)]

xs = [i for i , _ in enumerate(variance)]

```

```python
##imprimir os resultados

plt.plot(xs , variance , 'r-' , label='variance')
plt.plot(xs, bias_squared , 'b-', label='bias_squared')
plt.plot(xs, total_erro , 'black', label ='total erros')

plt.legend(loc= 6)


plt.show()
```
