```python
import matplotlib.pyplot as plt
```

```python
anos = [1950 , 1960 , 1970 , 1990 , 2000,2010]
gdp = [300.3 , 543.4 , 1075.9 , 2862.5 , 10283.7 , 1458.4]

plt.plot(anos , gdp , color = 'green' , marker ='o' , linestyle = 'solid')


plt.title('Nominal GPD')
plt.ylabel('Billions of $')
plt.show()
```
