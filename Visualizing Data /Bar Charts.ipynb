```python
import matplotlib.pyplot as plt 
%matplotlib notebook
```

```python
filmes = ['Batman ', 'Superman', 'MulherMavilha','Homen Formida']
mas_vistos = [100 , 200 , 300 , 150]
```

```python
filmes_x = [i+0.1 for i , _  in enumerate(filmes)]

plt.bar(filmes_x , mas_vistos)
plt.xticks([i+0.1 for i , _ in  enumerate(filmes)],filmes)
plt.ylabel('Filmes mais assistidos')
plt.title('Filmes ')
plt.show()
```
